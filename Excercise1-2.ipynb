```python
# Load libraries
import numpy as np
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
```

```python
# Load the Fashion MNIST dataset
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# Print number of classes and some examples of data
import matplotlib.pyplot as plt
num_classes = max(train_labels) + 1
print('Number of classes = {}'.format(num_classes))
plt.imshow(train_images[0], cmap='gray')
plt.figure()
plt.imshow(train_images[10000], cmap='gray')
plt.figure()
plt.imshow(train_images[20000], cmap='gray')
plt.figure()
plt.imshow(train_images[30000], cmap='gray')
plt.figure()
plt.imshow(train_images[40000], cmap='gray')
plt.figure()
plt.imshow(train_images[50000], cmap='gray')
plt.figure()
plt.imshow(train_images[59999], cmap='gray')
```

```python
# Data normalize and reshape
train_images = train_images.astype('float32') / 255
test_images = test_images.astype('float32') / 255

train_images = train_images.reshape(-1, 28*28)
test_images = test_images.reshape(-1, 28*28)
```

```python
# Model built
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=28*28))
model.add(Dense(10, activation='softmax'))
```

```python
# Model compile
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Train data
model.fit(train_images, train_labels, epochs=10, batch_size=64)
```

```python
# Evaluate model
test_loss, test_acc = model.evaluate(test_images,  test_labels)

print('\nTest accuracy: {}\nTest loss: {}'.format(test_acc, test_loss) )
```
